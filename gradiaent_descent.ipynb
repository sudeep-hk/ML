example1- A simple cost function

$f(x)=x^2+x+1$ 

```python
def f(x):
    return x**2+x+1
```

```python
# make data
x_1=np.linspace(start=-3,stop=3,num=100)
```

```python
# plot 
plt.figure(figsize=[10,5])

plt.subplot(1,2,1)
plt.xlim([-3,3])
plt.ylim([0,8])
plt.plot(x_1,f(x_1))

plt.subplot(1,2,2)
plt.plot(x_1,df(x_1),color='blue')
plt.show()
```
